```python
4. You have a list containing different data types: [1, "two", 3.0, [4,5], {"five": 5}, (6,)].
Write a Python function that processes this list and outputs
a. L=[12,’Python’,90.56,78,34,’Is’,65.90,’Easy’]
b. The sum of all integers and floats.
c. The concatenation of all strings.
d. The total number of elements in all sequences (lists, tuples, dictionaries)
```

```python
L=[12,'Python',90.56,78,34,'Is',65.90,'Easy']
print(L)
total = sum([n for n in L if isinstance(n, (int, float))])
print(f"the sum of the numeric element is :{total}")
```

```python
L = [12,'Python',90.56,78,34,'Is',65.90,'Easy']
concat = []
for item in L:
    if isinstance(item, str):
        concat.append(item)
print(concat)
```
