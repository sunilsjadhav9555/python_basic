```python
def countdown(n):
    if n == 0:
        print("Time's up!")
    else:
        print(n)
        countdown(n - 1)
```

```python
countdown(10)
```

```python
def factorial(n):
    if n  == 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
result = factorial(5)
print(result)
```

```python
lst = [1, 2, 3, 4, 5]
new_lst = list(map(lambda x: x ** 2, lst))
print(new_lst)
```

```python
lst = [1, -2, 3, -4, 5]
new_lst = list(filter(lambda x: x > 0, lst))
print(new_lst)
```

```python
temps_c = [0, 20, 37, 100]
temps_f = list(map(lambda c: (c * 9/5) + 32, temps_c))
print(temps_f)
```

```python
lst = [1, 2, 3, 4, 5, 6]
even_lst = list(filter(lambda x: (x % 2 == 0), lst))
print(even_lst)
```

```python
words = ["cat", "elephant", "dog", "tiger"]
long_words = list(filter(lambda w: len(w) > 4, words))
print(long_words)
```

```python
from functools import reduce

lst = [1, 2, 3, 4, 5]
product_lst = reduce(lambda x, y: x*y, lst)
print(product_lst)
```

```python
numbers = [5, 1, 8, 3, 7]
max_num = reduce(lambda a, b: a if a > b else b, numbers)
print(max_num)
```
