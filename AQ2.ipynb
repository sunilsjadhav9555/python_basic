```python
a = int(input('Please enter 1st  number')),
b = int(input('Please enter 2nd number')),
c = int(input('Please enter 3rd number')),

if a==b and a==c:
   print(f'All equls{a} {b} {c}'),
elif a >= b and a >= c:
   print(f'The first number ie {a} is greater than {b} and {c}'),
elif b >= a and b >= c:
   print(f'The second number ie {b} is greater than {a} and {c}'),
else :
  print(f'The Third number ie {c} is greater than {b} and {a}')
```
