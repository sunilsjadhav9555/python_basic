```python
def get_sum(lst):
    """
    this function returns the sum of all elements in the list
    """
    sum_num = 0
    for num in lst:
        sum_num = sum_num + num

    return sum_num
```

```python
result = get_sum(lst)
print(result)
```

```python
def computeHCF(a, b):
    """computing HCF of two numbers
    """
    smaller = b if a > b else a
    hcf = 1
    for i in range(1, smaller+1):
        if(a % i == 0) and (b % i == 0):
            hcf = i
        return hcf

num1 = 6
num2 = 36

print(f"H.C.F of {num1} and {num2} is : {computeHCF(num1, num2)}")
```

```python
def simple_interest(principal, rate, time=1):
    """
    calculate the simple interest
    formula : si = (p*r*t)/100
    """
    interest = (principal * rate * time)/100
    total_amount = principal + interest
    return interest, total_amount
```

```python
p = 100000
r = 10
t = 2
si, total = simple_interest(p, r, t)

print(f"simple interest = {si}")
print(f"total amount after {t} year = {total}")
```

```python
def compound_interest(p, r, t):
    """
    calculate compound interest.
    formula : A = P * (1 + R/(100*n))^(n*t)
    """
    total = p * (1 + r/100)**t
    ci = total - p
    return ci, total
```

```python
p = 10000
r = 10
t = 2

ci, total = compound_interest(p, r, t)

print(f"compound Interest = {ci}")
print(f"total amount after {t} years = {total}")
```
