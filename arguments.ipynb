```python
def book_flight(source, destination, date, seat_class):
    print(f"Booking from {source} to {destination} on {date} in {seat_class}")

book_flight("nyc","paris", "2025-07-20", "business")
```

```python
def greet(message, *names):
    for name in names:
        print(f"{message},{name}!")
```

```python
greet("hello","bob","sam","neel")
```

```python
def show_details(**kwargs):
    print("kwargs =", kwargs)
    for key, value in kwargs.items():
        print(f"{key} = {value}")
```

```python
show_details(name="alice", age =23,city="pune")
```
